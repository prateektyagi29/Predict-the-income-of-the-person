###  Detecting Outliers

```python
def outlier_detect(adult1):
    for i in adult1.describe().columns:
        Q1=adult1.describe().at['25%',i]
        Q3=adult1.describe().at['75%',i]
        IQR= Q3-Q1
        LTV = Q1-1.5*IQR
        UTV = Q3+1.5*IQR
        adult1[i]=adult1[i].mask(adult1[i]<LTV,LTV)
        adult1[i]=adult1[i].mask(adult1[i]>UTV,UTV)
    return adult1

#For detecting and removing the outliers. 
```

```python
adult1= outlier_detect(adult1)
```
